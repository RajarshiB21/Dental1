```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2

import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
import os
import seaborn as sns
from keras.applications.vgg16 import VGG16
```

```python
print(os.listdir("/content/drive/MyDrive/RAJ-AI/MISSING/"))

SIZE = 256  #Resize images

#Capture training data and labels into respective lists
train_images = []
train_labels = [] 
```

```python
for directory_path in glob.glob("/content/drive/MyDrive/RAJ-AI/MISSING/TRAIN/*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)       
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        train_images.append(img)
        train_labels.append(label)
```

```python
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
test_images = []
test_labels = [] 
for directory_path in glob.glob("/content/drive/MyDrive/RAJ-AI/MISSING/TEST/*"):
    fruit_label = directory_path.split("\\")[-1]
    print(fruit_label)
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        test_images.append(img)
        test_labels.append(fruit_label)
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)
```

```python
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
VGG_model = VGG16(weights='imagenet', include_top=False, input_shape=(SIZE, SIZE, 3))

#Make loaded layers as non-trainable. This is important as we want to work with pre-trained weights
for layer in VGG_model.layers:
	layer.trainable = False
    
VGG_model.summary()
```

```python
feature_extractor=VGG_model.predict(x_train)

features = feature_extractor.reshape(feature_extractor.shape[0], -1)
```

```python
X_for_training = features
```

```python
import xgboost as xgb
model = xgb.XGBClassifier()
model.fit(X_for_training, y_train)
```

```python
X_test_feature = VGG_model.predict(x_test)
X_test_features = X_test_feature.reshape(X_test_feature.shape[0], -1)
```

```python
prediction = model.predict(X_test_features)
#Inverse le transform to get original label back. 
#prediction = le.inverse_transform(prediction)
```
